```python
# Importing libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Using pivot table to effectively run .groupby function -- df.groupby('job_title_short).size()
df.pivot_table(index='job_title_short', aggfunc='size')


```

```python
# Making more use of the pivot table function to aggregate median salaries -- 
# df.groupby('job_title_short)['salary_year_avg'].median()
df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median')
```

```python
# But this time we break it down even further, by country
# This is not going to be a single line of code if trying to use groupby

df.pivot_table(values='salary_year_avg', index='job_title_short', columns='job_country', aggfunc='median')
```

```python
# Create a graph from the top 6 countries

# Extract top 6 countries based on frequency
top_countries = df['job_country'].value_counts().head(6).index

# Creating pivot table to compare median salaries between job title and country of posting
df_job_country_salary = df.pivot_table(
    values='salary_year_avg',
    index='job_country',
    columns='job_title_short',
    aggfunc='median'
)


df_job_country_salary = df_job_country_salary.loc[top_countries]

job_titles = ['Data Analyst', 'Data Scientist', 'Data Engineer']

df_job_country_salary = df_job_country_salary[job_titles]

df_job_country_salary.plot(kind='bar')
plt.xlabel('')
plt.ylabel('Yearly Salary')
plt.title('Median Salary by Country and Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
```
