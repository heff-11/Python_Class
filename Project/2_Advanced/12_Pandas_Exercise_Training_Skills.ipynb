```python
# Importing libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import ast

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
### Searching for trending job skills (which ones are most prevalent through time periods)

#   Filtering for specific job title creating a new dataframe

df_DS = df[df['job_title_short'] == "Data Scientist"].copy()
df_DS
```

```python
### Creating job posted month (by #) column for data frame coercing from job posted date column

df_DS['job_posted_month_no'] = df_DS['job_posted_date'].dt.month
df_DS 
```

```python
### Using explode() function to parse out individual skills from job postings

df_DS_explode = df_DS.explode('job_skills')

df_DS_explode

# Now we need to pivot this in order to use the created job month column and the exploded skills
    # Always run the code to check it works before assigning it a name
df_DS_pivot = df_DS_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

# Plot
    # Need to get rid of the way too many skills making this unreadable
# df_DS_pivot.plot(kind='line')

# Using .loc()
    # This sums the number of times each job skill was requested per month in a new row
df_DS_pivot.loc['Total'] = df_DS_pivot.sum()

#df_DS_pivot

# This is passing the sorting and index into the dataframe saying "Hey, sort this data by this"
    # After checking, reassign to the pivot table
df_DS_pivot = df_DS_pivot[df_DS_pivot.loc['Total'].sort_values(ascending=False).index]

# Get rid of the "Total row"
df_DS_pivot = df_DS_pivot.drop('Total')

df_DS_pivot

```

```python
df_DS_pivot.iloc[:, :5].plot(kind='line')

plt.title('Top Trending Job Skills for Data Scientists')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show
```
